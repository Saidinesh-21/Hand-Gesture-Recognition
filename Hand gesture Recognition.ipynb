```python
import numpy as np
import cv2
import os
import time
```

```python
hand_sign_images = []
target = []
sign_and_numbers = {}
sign_no = 0
```

```python
source = cv2.VideoCapture(0)
for i in range(5):
    image_no = 0
    sign = input("Please enter the sign : ")
    print("Starting in 5 seconds...")
    time.sleep(5)
    while 1:
        not_to_use, image = source.read()
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        x = 30
        y = 30
        w = 350
        h = 350
        hand_roi = gray[y:y+w, x:x+h]
        resized_image = cv2.resize(hand_roi, (100, 100))
        #normalized_image = resized_image/255
        hand_sign_images.append(resized_image)
        target.append(sign_no)

        cv2.putText(image, "Image No : " + str(image_no), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 100, 100), 2)
        image_no += 1
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
        cv2.imshow('Verzeo Project', image)
        cv2.imshow("Just Hand", hand_roi)
        key = cv2.waitKey(1)
        if image_no == 200:
            break
    
    
    sign_and_numbers[sign_no] = sign
    sign_no += 1
    cv2.destroyAllWindows()
source.release()
```

```python
hand_sign_images = np.resize(hand_sign_images, (2500, 100, 100, 1))
hand_sign_images
```

```python
from keras.utils import np_utils
target = np_utils.to_categorical(target)
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(hand_sign_images, target)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Activation, Dropout
from tensorflow.keraskeras.optimizers import SGD
from keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
```

```python
model.add(Conv2D(200, (3,3), input_shape=(100, 100, 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(100, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Dropout(0.3))
model.add(Flatten())

model.add(Dense(50, activation='relu'))
model.add(Dense(5, activation='softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cp = ModelCheckpoint('model-best2', verbose=0, save_best_only=True)
model.fit(X_train, Y_train, epochs = 3, callbacks=[cp], validation_split=0.2)
```

```python
from keras.models import load_model
model = load_model('model-best2')
```
